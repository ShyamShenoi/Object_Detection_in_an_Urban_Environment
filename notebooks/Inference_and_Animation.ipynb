# Inference and Video Animation
This notebook demonstrates how to load the exported trained model, pass a video or TFRecord dataset sequence into it, and generate an animated video of the bounding box detections.

```python
import os
import sys
sys.path.append('../scripts')
from inference_video import main as build_video

# Define Paths
LABELMAP_PATH = '../label_map.pbtxt'
MODEL_PATH = '../experiments/pretrained_model/saved_model' # Replace with exported model path
TF_RECORD_PATH = '../data/test/segment-12345.tfrecord' # Replace with your test tfrecord
CONFIG_PATH = '../experiments/pretrained_model/pipeline.config'
OUTPUT_PATH = 'animation.mp4'
```

## Generate Animation
We execute the exported `detect_fn` dynamically on frames of the TFRecord, visualizing the labeled boxes using `matplotlib.animation`.

```python
# If the files exist, this will build the mp4. 
# Note: running this requires standard TF2 environment and matplotlib.
if os.path.exists(MODEL_PATH) and os.path.exists(TF_RECORD_PATH):
    print('Starting Video Generation...')
    build_video(LABELMAP_PATH, MODEL_PATH, TF_RECORD_PATH, CONFIG_PATH, OUTPUT_PATH)
    print('Video saved to animation.mp4!')
else:
    print('Model or Testing data missing. Ensure you have exported the model first.')
```
